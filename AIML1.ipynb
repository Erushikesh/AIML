```python
#Part 1 â€“ Implement Basic Data Structures using Numpy, Pandas

# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Define a sample array
x = np.linspace(0.1, 10, 100)

# Perform sine operation
y_sin = np.sin(x)

# Perform logarithm operation
y_log = np.log(x)

# Perform exponential operation
y_exp = np.exp(x)

# Plot the results
plt.figure(figsize=(12, 6))
```

```python
# Plot sine
plt.subplot(1, 3, 1)
plt.plot(x, y_sin, label='sin(x)')
plt.title('Sine Function')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()

# Plot logarithm
plt.subplot(1, 3, 2)
plt.plot(x, y_log, label='log(x)', color='orange')
plt.title('Logarithm Function')
plt.xlabel('x')
plt.ylabel('log(x)')
plt.legend()

# Plot exponential
plt.subplot(1, 3, 3)
plt.plot(x, y_exp, label='exp(x)', color='green')
plt.title('Exponential Function')
plt.xlabel('x')
plt.ylabel('exp(x)')
plt.legend()

# Show plots
plt.tight_layout()
plt.show()
```

```python
# Part 2: Program 1
# First, you need to install the pandas library if you haven't already
!pip install pandas
# Import the pandas library
import pandas as pd
# Define the data
people = ['Kiran', 'Arun', 'Vijay', 'Varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]
# Create a DataFrame
data = {
    'Name': people,
    'Age': age,
    'Height': height,
    'Weight': weight
}
df = pd.DataFrame(data)
# Display the DataFrame
print(df)
```

```python
# 3. Scatter Plot: Age vs Weight
plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['Weight'], color='blue', marker='o')
plt.title('Scatter Plot of Age vs Weight')
plt.xlabel('Age (years)')
plt.ylabel('Weight (kg)')
plt.grid(True)
plt.show()
```

```python
# 4. Bar Chart: Height of Individuals
plt.figure(figsize=(8, 6))
plt.bar(df['Name'], df['Height'], color='green')
plt.title('Bar Chart of Individuals\' Heights')
plt.xlabel('Name')
plt.ylabel('Height (cm)')
plt.show()
```

```python
# 5. Histogram: Distribution of Age
plt.figure(figsize=(8, 6))
plt.hist(df['Age'], bins=5, color='orange', edgecolor='black')
plt.title('Histogram of Age Distribution')
plt.xlabel('Age (years)')
plt.ylabel('Frequency')
plt.show()
```
